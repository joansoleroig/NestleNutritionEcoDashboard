```python
import pandas as pd
import numpy as np
```

```python
# read csv
df1= pd.read_csv('nestle_es.csv', sep='\t')
df=df1
df.shape
```

```python
df.describe()
```

The main goal of this project will be to evaluate Nestlé's products that are commercialized in Spain, therefore, we will focus only on the products that contain a value in product_name_es, and get rid of the rest.

```python
# Get a list of all column names
cols = df.columns.tolist()

# Create a new list with columns that start with 'product_name'
product_name_cols = [col for col in cols if col.startswith('product_name')]

# Define a function to get the first non-null value from a list of columns
def get_first_non_null(row, columns):
    for col in columns:
        if pd.notnull(row[col]):
            return row[col]
    return np.nan

# Create the new 'name' column
df['product_name'] = df.apply(lambda row: row['product_name_es'] if pd.notnull(row['product_name_es']) else get_first_non_null(row, product_name_cols), axis=1)

df.shape
```

```python
# Check for missing values in 'name' column
missing_values = df['product_name'].isnull().sum()
print(f"Number of missing values in 'name' column: {missing_values}")
cols_to_drop = [col for col in cols if col.startswith('product_name') and col != 'product_name']
df = df.drop(columns=cols_to_drop)
# delete columns starting with 'p
df.shape
```

```python
#drop all columns containing 'generic_name'
generic_name_cols = [col for col in cols if 'generic_name' in col]
df = df.drop(columns=generic_name_cols)
df.shape
```

```python
cols = df.columns.tolist()
print(cols)
```

```python
# delete columns 'abbreviated_column_name_fr', 'code', 'product_version_id', also columns which have tags as we will work with the tagged ones.
cols_to_drop = ['abbreviated_product_name_fr', 'code', 'producer_version_id', 'brands', 'packaging', 'brand_owner', 'categories', 'labels', 'countries', 'stores', 'obsolete', 'obsolete_since_date', 'origins', 'origins_tags', 'origin_es', 'origin_fr', 'manufacturing_places','manufacturing_places_tags', 'emb_codes', 'emb_codes_tags' ]
df = df.drop(columns=cols_to_drop)
df.shape
```

```python
# Strip the string in 'quantity' column until the first 'g', do not care about warnings
df['quantity'] = df['quantity'].apply(lambda x: x.split('g', 1)[0] if isinstance(x, str) else x)
```

```python
# Dealing with ingredient columns
ingredient_cols = [col for col in cols if col.startswith('ingredients')]
df['ingredients'] = df.apply(lambda row: row['ingredients_text_es'] if pd.notnull(row['ingredients_text_es']) else get_first_non_null(row, ingredient_cols), axis=1)
cols_to_drop = [col for col in cols if col.startswith('ingredients') and col != 'ingredients']
df = df.drop(columns=cols_to_drop)
df.shape
```

```python
cols_to_drop = ['allergens', 'traces', 'no_nutrition_data', 'energy-kj_unit', 'energy-kj_prepared_value', 'energy-kj_prepared_unit', 'energy-kcal_unit', 'energy-kcal_prepared_value', 'energy-kcal_prepared_unit', 'fat_prepared_value', 'fat_prepared_unit', 'saturated-fat_prepared_value', 'saturated-fat_prepared_unit', 'carbohydrates_prepared_value', 'carbohydrates_prepared_unit', 'sugars_prepared_value', 'sugars_prepared_unit', 'fiber_prepared_value', 'fiber_prepared_unit']
df = df.drop(columns=cols_to_drop)
df.shape
```

```python
# drop all columns ending with '_prepared_value' or '_prepared_unit'
cols_to_drop = [col for col in df.columns if col.endswith('_prepared_value') or col.endswith('_prepared_unit')]

# Only keep columns in cols_to_drop that are actually in the dataframe
cols_to_drop = [col for col in cols_to_drop if col in df.columns]

df = df.drop(columns=cols_to_drop)
df.shape
```

```python
cols_to_drop = ['alcohol_value', 'alcohol_unit', 'arachidic-acid_value', 'arachidic-acid_unit', 'monounsaturated-fat_value', 'monounsaturated-fat_unit', 'polyunsaturated-fat_value', 'polyunsaturated-fat_unit']
df = df.drop(columns=cols_to_drop)
df.shape
```

```python
#drop columns with more than 99% missing values
missing_values = df.isnull().sum()
cols_to_drop = missing_values[missing_values > 0.97 * df.shape[0]].index
df = df.drop(columns=cols_to_drop)
df.shape
```

```python
# if the value of the column 'salt_unit' is 'mg', multiply the value in 'salt_value' by 0.001
df.loc[df['salt_unit'] == 'mg', 'salt_value'] = df['salt_value'] * 0.001
```

```python
df.to_csv('nestle_es_cleaned.csv', index=False)
```
